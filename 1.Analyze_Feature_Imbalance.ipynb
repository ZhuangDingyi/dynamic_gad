```python
# Based on the results from the slides
import matplotlib.pyplot as plt
import numpy as np

# Data from the table
percentages = [10, 20, 30, 40, 50, 60, 70, 80, 90]
macro_f1 = [0.8716, 0.7258, 0.7816, 0.7331, 0.7473, 0.8576, 0.4896, 0.4916, 0.0455]
auc = [0.9131, 0.9301, 0.8278, 0.9058, 0.8846, 0.8869, 0.5, 0.5, 0.5]

# Plotting the trends
plt.figure(figsize=(10, 6))

# Plot Macro-F1
plt.plot(percentages, macro_f1, marker='o', linestyle='-', color='b', label='Macro-F1')

# Plot AUC
plt.plot(percentages, auc, marker='o', linestyle='-', color='g', label='AUC')

# Adding titles and labels
plt.title('Trend of Macro-F1 and AUC According to Different Percentages of External Nodes')
plt.xlabel('Percentage of External Nodes')
plt.ylabel('Scores')
plt.xticks(np.arange(10, 100, 10))
plt.legend()
plt.grid(True)
plt.show()

```
